```python
def convert_data(custom_data):
	total_data={
        "images":[],
        "annotations":[],
        "categories":[
            {"supercategory":"indoor" , "id":1 , "name":"door"}
        ]
    }
	
	prev_img =""
	img_cnt = -1
	all_cnt = 0
	for anno in custom_data:
		crn_img = anno['image']	

		if prev_img != crn_img:
			img_cnt+=1
			total_data['images'].append(
				{"file_name": crn_img , "height": 1024 , "width":2048 , "id":img_cnt}
			)
		
		#anno
		for single_door_u , single_door_v  in zip(anno['u'] , anno['sticks_v'] ):
			box_w = abs(single_door_u[0] - single_door_u[1]) * 2048
			box_h = abs(single_door_v[3] - single_door_v[0]  ) * 1024
			bbox = [single_door_u[0]  *2048 , single_door_v[0] *1024 ,  box_w , box_h]


			anno_data={
					"image_id": img_cnt , 
					"file_name":crn_img,
					"id":all_cnt,
					"iscrowd":0,
					"category_id":1,                  
					"bbox": bbox,
					"area": box_w * box_h
				}   

			all_cnt+=1
			total_data['annotations'].append(anno_data)

	return total_data
	pass


import json
import os

# Source
json_path = os.path.join(os.getcwd(), 'anno', 'train_visiable_all.json')
f= open(json_path)
custom_anno = json.loads(f.read())
f.close()
output = convert_data(custom_anno)

# Output
output_path = os.path.join(os.getcwd() , "anno" , "coco_train_visiable_all.json")
with open(output_path, 'w') as f:
    json.dump( output , f)
```
